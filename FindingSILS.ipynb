# Finding SILS


```python
import loopset
```

```python
data = loopset.read_from_pickle('temp.pickle')

#Visualise original pickle graph
nx.draw_networkx(data)
plt.show()
```

```python
cycleGraphs = loopset.cycleGraph(data)

#Visualise cycle graphs
fig = plt.figure()
for graph in cycleGraphs:
    ax = fig.add_subplot()
    nx.draw_networkx(graph, pos=nx.spring_layout(graph), ax=ax)
plt.show()
```

```python
#loopset.build_sils(cycleGraphs)

for graph in cycleGraphs:
    sils = loopset.SILS(graph)
    for entry in sils:
        print (entry)
```
