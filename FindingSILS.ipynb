# Finding SILS

Usage for reading in a causal map and finding the SILS

```python
import loopset
```

```python
data = loopset.read_from_pickle('test.pickle')
sils = loopset.build_sils(data)

sils
```

### Verify Visually

```python
cycleGraphs = loopset.cycleGraph(data)

fig = plt.figure()
for graph in cycleGraphs:
    ax = fig.add_subplot()
    nx.draw_networkx(graph, pos=nx.spring_layout(graph), ax=ax)
plt.show()
```
